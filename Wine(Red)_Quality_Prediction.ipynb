Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
wine_dataset = pd.read_csv('/content/winequality-red.csv')
```

```python
wine_dataset.head()
```

```python
wine_dataset["quality"].value_counts()
```

```python
wine_dataset.shape
```

```python
wine_dataset.describe()
```

```python
wine_dataset.isnull().sum()
```

```python
wine_dataset["quality"] = wine_dataset["quality"].apply(lambda y: 1 if y>=7 else 0)
```

```python
wine_dataset.head()
```

```python
wine_dataset.groupby('quality').mean()
```

Data Spliting

```python
X = wine_dataset.drop(columns='quality', axis=1)
Y = wine_dataset['quality']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=6969)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Data Train

```python
classifier = RandomForestClassifier()
```

```python
classifier.fit(X_train, Y_train)
```

Evaluation

```python
X_train_prediction = classifier.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(f"Accuracy with train data {train_data_accuracy}")
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(f"Accuracy with test data {test_data_accuracy}")
```
